```python
# 함수화

def multi(values):
    for i in range(1,len(values)-1):
        if values[i]=='*':
            values[i+1]=int(values[i-1])*int(values[i+1])

    while '*' in values:
        del values[values.index('*')-1:values.index('*')+1]
    return values

def plus(values):
    for i in range(1,len(values)-1):
        if values[i]=='+':
            values[i+1]=int(values[i-1])+int(values[i+1])

    while '+' in values:
        del values[values.index('+')-1:values.index('+')+1]
    return values

def minus(values):
    for i in range(1,len(values)-1):
        if values[i]=='-':
            values[i+1]=int(values[i-1])-int(values[i+1])

    while '-' in values:
        del values[values.index('-')-1:values.index('-')+1]
    return values

def multi_plus_minus(values):
    copy_value = values[:]
    copy_value=multi(copy_value)
    copy_value=plus(copy_value)
    copy_value=minus(copy_value)
    return copy_value[0]

def multi_minus_plus(values):
    copy_value = values[:]
    copy_value=multi(copy_value)
    copy_value=minus(copy_value)
    copy_value=plus(copy_value)
    return copy_value[0]

def minus_plus_multi(values):
    copy_value = values[:]
    copy_value=minus(copy_value)
    copy_value=plus(copy_value)
    copy_value=multi(copy_value)
    return copy_value[0]

def minus_multi_plus(values):
    copy_value = values[:]
    copy_value=minus(copy_value)
    copy_value=multi(copy_value)
    copy_value=plus(copy_value)
    return copy_value[0]

def plus_minus_multi(values):
    copy_value = values[:]    
    copy_value=plus(copy_value)
    copy_value=minus(copy_value)
    copy_value=multi(copy_value)
    return copy_value[0]

def plus_multi_minus(values):
    copy_value = values[:]    
    copy_value=plus(copy_value)
    copy_value=multi(copy_value)
    copy_value=minus(copy_value)
    return copy_value[0]

def solution(expression):

    nums = "0123456789"
    values = []
    adds = []
    for i in range(len(expression)):
        if i==0 or expression[i] not in nums or values[-1][-1] not in nums:
            values.append(expression[i])
        else:
            values[-1]+=expression[i]
    adds.append(abs(multi_plus_minus(values)))
    adds.append(abs(multi_minus_plus(values)))
    adds.append(abs(plus_multi_minus(values)))
    adds.append(abs(plus_minus_multi(values)))
    adds.append(abs(minus_plus_multi(values)))
    adds.append(abs(minus_multi_plus(values)))  
    
    return max(adds)
expression = "50*6-3*2"  
solution(expression)

## 열심히 만들었으나.....초기값인 value가 원상태를 유지하지 않는 점을 고려하지 않아..다시 작성 필요--수정완료..
```

```python
## Eval이라는 함수를 새로 배웠다! 이것을 적용해보자

expression = "50*6-3*2"  
nums = "0123456789"
values = []
adds = []
for i in range(len(expression)):
    if i==0 or expression[i] not in nums or values[-1][-1] not in nums:
        values.append(expression[i])
    else:
        values[-1]+=expression[i]

def cal(values,a):
    for i in range(1,len(values)-1):
        if values[i]==a:
            values[i+1]=eval(values[i-1]+a+values[i+1])

    while a in values:
        del values[values.index(a)-1:values.index(a)+1]
    return values
cal(values,'*')

# 반복되는 식이 너무 많다
```
